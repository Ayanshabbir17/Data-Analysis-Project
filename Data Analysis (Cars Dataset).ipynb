# Data Cleaning

Find all the null value in the dataset.If there is any null value in any column,then fill it with the mean of that column and for non-numerical column fill it with mode of that column.

```python
car.isnull().sum()
```

In 'cylinder' column only 2 null values are there.

```python
#To replace null values in cylinder column
car['Cylinders'].fillna(car['Cylinders'].mean(), inplace=True)
```
